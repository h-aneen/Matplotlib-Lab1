## Q2: Create pie chart with equal explode values.

```python
mylabels = ["Tomatoes", "Mangoes", "Oranges", "Apples"]
  
# list of int
y = [25,55,60,40]
myexplode = [0,0,0,0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
mylabels = ["Tomatoes", "Mangoes", "Oranges", "Apples"]
y = [25,55,60,40]
df=pd.DataFrame(list(zip(mylabels,y)),columns=['fruit','number'])

plt.scatter(df['fruit'],df['number'],alpha =0.4,s=150) 
plt.xlabel('fruit',fontsize = 16)
plt.ylabel('number of people',fontsize = 16)
plt.title('Distribution ',color='r',loc="left")
```

## Q5: Create three line charts with Legends at left and colors

```python
y1 = [2, 3, 4.5]
  
# Y-axis values 
y2 = [1, 1.5, 5]

y3 = [1, 1.5, 3]
  
# Function to plot  
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
  
# Function add a legend  
plt.legend(['y1', 'y2','y3'], loc ='upper left')
  
# function to show the plot
plt.show()



```

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

plt.subplot(1,3,1)
plt.plot(y1)
plt.legend(['y1', 'y2','y3'], loc ='upper left')
plt.subplot(1,3,2)
plt.scatter(df['fruit'],df['number'],alpha =0.4,s=150) 
plt.xlabel('fruit',fontsize = 16)
plt.ylabel('number of people',fontsize = 16)
plt.title('Distribution ',color='r',loc="left")
plt.subplot(1,3,3)
mylabels = ["Tomatoes", "Mangoes", "Oranges", "Apples"]
y = [25,55,60,40]
myexplode = [0,0,0,0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```
